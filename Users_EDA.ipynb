```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
users = pd.read_csv('users.csv')
```

```python
pd.set_option('max_columns', None)
users.columns = users.columns.str.replace('.', '_')
users.head()
```

```python
users.isnull().sum()
```

```python
users.info()
```

```python
sns.heatmap(users.isnull(), cbar = False, cmap = 'viridis')
```

```python
for col in users.columns:
    print('{} : {}'.format(col,users[col].unique()))
```

```python
gen = np.repeat('r', 10000)
for index,i in enumerate(users['gender']):
    if i == "male":
        gen[index] = 'g'
users.plot(xlabel = 'latidute', ylabel = 'longitude', x = 'location_coordinates_latitude', y = 'location_coordinates_longitude', color = gen, kind = 'scatter')
```

```python
sns.histplot(users.dob_age, bins = 50)
```

```python
pass_len = np.zeros(10000)
for index, i in enumerate(users['login_password']):
    pass_len[index] = len(i)
```

```python
sns.histplot(pass_len, bins = 10,).set(xlabel='password_length')
```

```python
uname_len = np.zeros(10000)
for index, i in enumerate(users['login_username']):
    uname_len[index] = len(i)
sns.histplot(uname_len, bins = 10,).set(xlabel='username_length')
```

```python
nat = ['DE' 'FI' 'DK' 'NO' 'BR' 'GB' 'TR' 'IE' 'AU' 'ES' 'IR' 'FR' 'CA' 'NZ'
 'CH' 'NL' 'US']
sns.histplot(users.nat, bins = len(nat),)
```

```python
ax = sns.countplot(x=users.location_timezone_offset)
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()
plt.show()
```
